Linear Model \
y = wx + b \
y = price of car \
price = wx + b

```python
def estimate_price(enginesize, w, b):
    return w * enginesize + b
```

Lets guess the w and b are 10 and 20

```python
w = 10
b = 20
```

```python
enginesizes = CarPrice_df.enginesize
estimated_price = estimate_price(enginesizes, w, b)
```

```python
plt.plot(enginesizes, estimated_price, 'r-o');
plt.xlabel('Engize Size');
plt.ylabel('Price');
```

```python
target = CarPrice_df.price

plt.plot(enginesizes, estimated_price, 'r', alpha=0.9);
plt.scatter(enginesizes, target, s=8,alpha=0.8);
plt.xlabel('Engine Size');
plt.ylabel('Price')
plt.legend(['Estimate', 'Actual']);
```

Estimate isnt good enough and therefore needs further needs to evaluated with new values

```python
def try_parameters(w, b):
    enginesizes = CarPrice_df.enginesize
    target = CarPrice_df.price
    
    estimated_price = estimate_price(enginesizes, w, b)
    
    plt.plot(enginesizes, estimated_price, 'r', alpha=0.9)
    plt.scatter(enginesizes, target, s=8,alpha=0.8)
    plt.xlabel('EngineSize')
    plt.ylabel('Price')
    plt.legend(['Estimate', 'Actual'])
```

```python
try_parameters(115, 100)
```

Loss Cost Function

1. Difference between the predicted value and actual value

2. Average out the sum of the squared difference between predicted and actual values

3. RMSE can be calculated by taking the square root of above mentioned Mean Squared Errors (MSE) / L2 Loss. The effect of each error on RMSE is proportional to the size of the squared error. It represents the square root of the second sample moment of the deviation between predicted model values and actual values.

More details:

https://medium.com/analytics-vidhya/common-loss-functions-in-machine-learning-for-a-regression-model-27d2bbda9c93

```python
import numpy as np
```

```python
def rmse(targets, predictions):
    return np.sqrt(np.mean(np.square(targets - predictions)))

w = 115
b = 100
```

```python
try_parameters(w, b)
```

```python
targets = CarPrice_df['price']
predicted = estimate_price(CarPrice_df.price, w, b)
```

```python
rmse(targets, predicted)
```

```python
def try_parameters(w, b):
    enginesizes = CarPrice_df.enginesize
    target = CarPrice_df.price
    predictions = estimate_price(enginesizes, w, b)
    
    plt.plot(enginesizes, predictions, 'r', alpha=0.9);
    plt.scatter(enginesizes, target, s=8,alpha=0.8);
    plt.xlabel('EngineSize');
    plt.ylabel('Price')
    plt.legend(['Prediction', 'Actual']);
    
    loss = rmse(target, predictions)
    print("RMSE Loss: ", loss)
```

```python
try_parameters(115, 100)
```
